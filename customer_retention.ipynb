Following interpretation could be made from the graphical analysis of numerical features

The customers have been using mobile  internet for shopping.The device used for shopping was found to be smartphones, followed by laptops.
3.Windows/windows mobile is being used as OS with google chrome as a browser.
4.Most of the customer are using debit/credit cards for payment,followed by COD and e-wallet being used the least.
5.Most common reason for leaving the cart without shopping is when better alternate offer is found on some other site.
6.The content must be clear and understandable. Also the information of similar product for product comparison is also very important.
7.Ease of navigation in website,along with loading and processing speed is very crucial for customers.User friendly interface is another important feature to be taken into consideration
8.Customer care facility,solving queries,privacy and responsiveness are another important features important for customers.
9.Monetary benefit, return/replacement policy, access to loyalty program, wide variety of product has been found to be strongly preferred.


# Bi-variate analysis

```python
#cat-cat columns
sns.countplot(df['preferred payment Option'],palette='coolwarm',hue='Gender',data=df)
```

Majority of the customer, irrespective of gender, prefer Credit/Debit cards as a payment options. However COD is more preferred in Females as compared to males

```python
sns.countplot(df['Age'],palette='coolwarm',hue='Gender',data=df)
```

Majority of the customer doing online shopping are between age group 20-50.There are very less customers below 20 years and more than 50 years in both males and females but the ratio of males to females in above 50 age group is less as compared to below 20 age group.

```python
sns.countplot(df['abandon_frequency'],palette='coolwarm',hue='Gender',data=df)
```

The frequency of abandoning a cart is greater in females than males.

```python
sns.countplot(df['device'],palette='coolwarm',hue='Gender',data=df)
```

Most of the people shop via smartphones followed by laptops.The ratio of males to females is almost equal in using Desktop.Females have not been found using tablet.

```python
sns.countplot(df['time_taken'],palette='coolwarm',hue='Gender',data=df)
```

Time taken to purchase a product is greater in females(more than 15 mins)as compared to males.Also males generally take 6-10 mins to purchase a product.

# Multivariate analysis

```python
cor=df.corr()
cor
```

```python
plt.figure(figsize=[22,12])
sns.heatmap(cor,annot=True)
plt.title('correlation matrix')
plt.show
```

1.A strong and obvious positive correlation was found between customer trust and retailer's empathy towards customer.Trust will be built in response of retailers response to queries.
2.There is strong correlation between monetary benefit and enjoyment which is quite understandable.
3.Also customers enjoy more with ease of navigation, faster loading and processing speed and user friendly interface.
4.Majority of the customer strongly agreed on the importance of display of content on the website.The content must be easy to read and understand.
Also, Information on similar product for product comparison, Complete information on listed seller and product, clear relevant information on listed products plays important role.
